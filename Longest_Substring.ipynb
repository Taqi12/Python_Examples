## Longest Substring Without Repeating Characters

```python
def longest_substring(s):
    seen = {}
    start = 0
    length = 0

    for i in range(len(s)):
        if s[i] in seen and start <= seen[s[i]]:
            start = seen[s[i]] + 1
        else:
            length = max(length, i - start + 1)
        seen[s[i]] = i
    return length
```

```python
# calling the function
longest_substring("abcabcbb")
```
